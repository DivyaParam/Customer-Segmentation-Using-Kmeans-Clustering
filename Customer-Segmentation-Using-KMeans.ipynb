```python
#Creating a new column for Age calcualtion of the Customer by year of birth 
data_frame['Age_on_2014'] = 2014 - data_frame['Year_Birth']
data_frame
```

```python
#Adding a new column - Total - Spend which is the sum of money spent on fruits, meats, fish products, sweet products, and gold products
data_frame['Spent'] = data_frame['MntFruits']+ data_frame['MntMeatProducts']+ data_frame['MntFishProducts']+data_frame['MntSweetProducts']+ data_frame['MntGoldProds']
data_frame
```

```python
#Adding another column - Living With based on the Marital Status of each individual
data_frame['Living_With'] = data_frame['Marital_Status'].replace({"Married":"Partner", "Together":"Partner", "Absurd":"Alone", "Widow":"Alone", "YOLO":"Alone", "Divorced":"Alone", "Single":"Alone"})
data_frame
```

```python
#To calculate the total children in the household by creating a new column Children
data_frame['Children'] = data_frame['Kidhome'] + data_frame['Teenhome']
data_frame
```

```python
#Adding another new column Family_Size
data_frame['Family_Size'] = data_frame['Living_With'].replace({'Alone':1,'Partner':2}) + data_frame['Children']
data_frame
```

```python
#Creating a new column Parent to indicate parenthood status
data_frame['Is_Parent'] = np.where(data_frame.Children > 0,1,0)
data_frame
```

```python
#Simplyfying the Education details 
data_frame['Education'] = data_frame['Education'].replace({"Basic":"Undergraduate", "2n Cycle":"Undergraduate", "Graduation":"Graduate", "Master":"Postgraduate", "PhD":"Postgraduate"})
data_frame
```

```python
#Dimensionality reduction 
#Creating a copy of the data frame to employ dimensionality reduction
data_frame1 = data_frame.copy()
```

```python
#Using PCA for dimensionality reduction
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(data_frame1)
scaled_data_frame1 = pd.DataFrame(scaler.transform(data_frame1), columns = data_frame1.columns)
scaled_data_frame1
```

```python
from sklearn.decomposition import PCA
pca = PCA(random_state = 42, svd_solver = 'full')
pca.fit(scaled_data_frame1)
cumsum = np.cumsum(pca.explained_variance_ratio_)
d = np.argmax(cumsum >= 0.95) + 1
print("The minimum dimensions requird to preserve a 95% variance", d)
```

```python
pca = PCA(n_components = 19, random_state = 42, svd_solver = 'full')
pca.fit(scaled_data_frame1)
data_frame1_reduced = pd.DataFrame(pca.transform(scaled_data_frame1), columns = (['col1', 'col2', 'col3', 'col4',
                                                                  'col5', 'col6', 'col7', 'col8',
                                                                  'col9', 'col10', 'col11', 'col12',
                                                                  'col13', 'col14', 'col15', 'col16',
                                                                  'col17', 'col18', 'col19']))
data_frame1_reduced
```

```python
#Applying KMeans Clustering using KElbow Visualizer
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
print("Elbow Method to determine the number of clusters to be formed:")
elbow = KElbowVisualizer(KMeans(), k = 10)
elbow.fit(data_frame1_reduced)
elbow.show()
```

```python
from sklearn.metrics import silhouette_score

kmeans_per_k = [KMeans(n_clusters=k, n_init=10, random_state=42).fit(data_frame1_reduced)
                for k in range(2, 11)]

silhouette_scores = [silhouette_score(data_frame1_reduced, model.labels_)
                     for model in kmeans_per_k[1:]]
plt.figure(figsize=(8, 3))
plt.plot(range(2, 10), silhouette_scores, "bo-")
plt.xlabel("$k$")
plt.ylabel("Silhouette score")
plt.grid(True)
plt.show()
```

```python
#Printing Silhouette score for each cluster

cluster_range = range(2, 10)
for i, score in zip(cluster_range, silhouette_scores):
  print(f"Silhouette Score for {i} Clusters:", score)
```

```python
from sklearn.metrics import silhouette_samples
from matplotlib.ticker import FixedLocator, FixedFormatter

plt.figure(figsize=(11, 10))

for k in (2, 3, 4, 5):
    plt.subplot(4, 2, k - 1)

    y_pred = kmeans_per_k[k - 1].labels_
    silhouette_coefficients = silhouette_samples(data_frame1_reduced, y_pred)

    padding = len(data_frame1_reduced) // 30
    pos = padding
    ticks = []
    for i in range(k):
        coeffs = silhouette_coefficients[y_pred == i]
        coeffs.sort()

        color = plt.cm.Spectral(i / k)
        plt.fill_betweenx(np.arange(pos, pos + len(coeffs)), 0, coeffs,
                          facecolor=color, edgecolor=color, alpha=0.7)
        ticks.append(pos + len(coeffs) // 2)
        pos += len(coeffs) + padding

    plt.gca().yaxis.set_major_locator(FixedLocator(ticks))
    plt.gca().yaxis.set_major_formatter(FixedFormatter(range(k)))
    if k in (3, 5):
        plt.ylabel("Cluster")

    if k in (5, 6):
        plt.gca().set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
        plt.xlabel("Silhouette Coefficient")
    else:
        plt.tick_params(labelbottom=False)

    plt.axvline(x=silhouette_scores[k - 2], color="red", linestyle="--")
    plt.title(f"$k={k}$")

plt.show()
```

```python
#Apply KMEans clustering for K = 2
kmeans = KMeans(n_clusters= 2, random_state=42)
cluster_labels = kmeans.fit_predict(data_frame1_reduced)
data_frame1['Cluster'] = cluster_labels
data_frame1.to_excel('Clustered_data.xlsx', index = False)
data_frame1.head()
```
